```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
%matplotlib inline
```

```python
#x from 0 to 30
x = 30 * np.random.random((100, 1))

# y = a*x + b with noise
y = 0.5 * x**3.5 + 1.0 + np.random.normal(size=x.shape)*5000
```

```python
# create a linear regression model
#model = LinearRegression()
#model.fit(x, y)
```

```python
clf = SVR(kernel='poly', C=100, gamma='auto', degree=3, epsilon=.1, coef0=1)
model = clf.fit(x, y.ravel()) 
```

```python
# predict y from the data
x_new = np.linspace(0, 30, 100)
y_new = model.predict(x_new[:, np.newaxis])
```

```python
# plot the results
fig = plt.figure(figsize=(12, 9))
fig.suptitle('Regression example', fontsize=30)
ax = plt.axes()
ax.scatter(x, y)
ax.plot(x_new, y_new)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')


plt.show()


```
