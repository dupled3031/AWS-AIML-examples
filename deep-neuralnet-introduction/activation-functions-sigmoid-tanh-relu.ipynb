```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)  
    return s,ds
```

```python
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return t,dt
```

```python
def relu(X, alpha=0.1):
    dx = np.ones_like(x)
    dx[x < 0] = alpha
    return np.maximum(alpha*X,X), dx
```

```python
def softmax(X):
    expo = np.exp(X)
    expo_sum = np.sum(np.exp(X))
    return expo/expo_sum
```

```python
def plotfunction(x, y, dy, label):

    fig, ax = plt.subplots(figsize=(9, 5))
    fig.suptitle(label, fontsize=32)
    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.plot(x,y, color="#307EC7", linewidth=3, label=label)
    ax.plot(x,dy, color="#9621E2", linewidth=3, label="derivative", linestyle=":")
    ax.legend(loc="upper right", frameon=False)
    fig.show()    
```

```python
x=np.arange(-6,6,0.01)
```

```python
plotfunction(x=x, y=sigmoid(x)[0], dy=sigmoid(x)[1], label="sigmoid")
```

```python
x=np.arange(-3,3,0.01)
plotfunction(x=x, y=tanh(x)[0], dy=tanh(x)[1], label="tanh")
```

```python
plotfunction(x=x, y=relu(x)[0], dy=relu(x)[1], label="leaky relu")
```
